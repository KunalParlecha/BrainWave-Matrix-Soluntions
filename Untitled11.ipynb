```python
df["CustomerID"]=df["CustomerID"].fillna(0).astype(int)
```

```python
#Here i am converting the data type from float to integer 
```

```python
df["Revenue"]=df["Quantity"]*df["UnitPrice"]
df["Revenue"].sum()
```

TOTAL REVENUE 

```python
top_products= df.groupby("Description")["Revenue"].sum().sort_values(ascending=False).head(10)
print(top_products)
```

REVENUE OF TOP 10 PRODUCTS

```python
sales_by_country=df.groupby("Country")["Revenue"].sum().sort_values(ascending=False).head(10)
print(sales_by_country)
```

SALES OF EACH COUNTRY 


```python
df["InvoiceDate"]=pd.to_datetime(df["InvoiceDate"])
df["Month"]=df["InvoiceDate"].dt.to_period("M")
```

```python
monthly_sales=df.groupby("Month")["Revenue"].sum()
print(monthly_sales)
```

```python
monthly_sales.plot(kind="line",figsize=(12,5),title="Monthly Sales Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

top_countries = df.groupby("Country")["Revenue"].sum().nlargest(5)

plt.figure(figsize=(13,13))  # Ensure this line is correct
top_countries.plot(kind="pie", autopct='%1.1f%%', startangle=90, cmap="viridis")
plt.title("Sales Contribution by Country")
plt.show()
```
